```python
# Lecture 52, notes by Akhona Njeje.
# Date 13 February 2023.
# Seaborn = Statistical plotting library.
# Works well with Pandas library.
# Consult documentation http://seaborn.pydata.org/ for other  tools.

# Regression , Linear Model plot.

import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex',row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time',hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',hue='sex',
          aspect=0.6,size=8)
```

```python
# Part 2 : STyle.

sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```
