```python
# Lecture 47, notes by Akhona Njeje.
# Date 13 February 2023.
# Seaborn = Statistical plotting library.
# Works well with Pandas library.
# Consult documentation http://seaborn.pydata.org/ for other  tools.

# Matrix Plots.

import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True)
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap = 'magma')
```

```python
sns.heatmap(fp, cmap = 'magma', linewidths = 3)
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp, cmap="coolwarm")
```

```python
sns.clustermap(fp, cmap="coolwarm", standard_scale=1)
```

```python
# Part 2 : Grids.

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)   # Install Matplot as plt.
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
```

```python
z = sns.FacetGrid(data=tips, col='time', row='smoker')
z.map(sns.displot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
